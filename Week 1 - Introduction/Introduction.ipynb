### Implementation and Results
To display the current date/time, we import the **time** module that allows the time to be represented as strings, numbers or even objects.

```python
# First Python Program
import time
print("Hello World!")
print("The current time is", time.ctime())
```

To display the waveform, we import the following libraries in our code:
1.   **matplotlib** contains methods for plotting and visualization of data.
2.   **numpy** contains methods for array creation and manipulation.



```python
# Graphical Output
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2 * np.pi * t)
plt.plot(t,s)
plt.title('A Simple Plot of Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
```
